```python
import control as ctl
import matplotlib.pyplot as plt
```

```python
# define a second-order system
sys = ctl.TransferFunction([1], [1, 2, 1])  # 1/(sÂ²+2s+1)
# braw Bode plot
ctl.bode_plot(sys, dB=True)
plt.show()
```

## Typical Elements

1. Proportional element

```python
sys1 = ctl.TransferFunction([10],[1], name="sys1")  # 5
ctl.bode_plot(sys1, dB=True)
plt.show()
```

2. Inertial element

```python
tau = 0.5
sys2 = ctl.TransferFunction([1], [tau, 1], name="sys2")  # 1/(0.5s+1)
ctl.bode_plot(sys2, dB=True)
plt.show()
```

3. Intergral element

```python
sys3 = ctl.TransferFunction([1], [1, 0], name="sys3")  # 1/(s)
ctl.bode_plot(sys3, dB=True)
plt.show()
```

4. Diffrential element

```python
sys4 = ctl.TransferFunction([1, 0], [1])  # s
ctl.bode_plot(sys4, dB=True)
plt.show()
```

5. First order differential element

```python
sys5 = ctl.TransferFunction([tao, 1], [1])  # s + 1
ctl.bode_plot(sys5, dB=True)
plt.show()
```

6. Oscillating element

```python
omega = 1 / tau
xi = 0.5
sys6 = ctl.TransferFunction([omega**2], [1, 2*xi*omega, omega**2], name="sys6")  # ÏÂ²/(sÂ²+2Î¾Ïs+ÏÂ²)
ctl.bode_plot(sys6, dB=True)
plt.show()
```

7. Second order differential element

```python
sys7 = ctl.TransferFunction([1, 2*xi*omega, omega**2], [omega**2], name="sys7")  # (sÂ²+2Î¾Ïs+ÏÂ²)/ÏÂ²
ctl.bode_plot(sys7, dB=True)
plt.show()
```

8. Delay element

```python
pade_order = 3
num, den = ctl.pade(tau, pade_order)  # åå­/åæ¯ç³»æ°
sys8 = ctl.TransferFunction(num, den, name="sys8")
ctl.bode_plot(sys8, dB=True)
plt.show()
```

9. Non-minimum phase element

```python
sys9 = ctl.TransferFunction([1], [tau, -1], name="sys9")  # éæå°ç¸ä½åä»¶ 1/(Ïs - 1)
ctl.bode_plot(sys9, dB=True)
plt.show()
```

```python
response = ctl.frequency_response([sys1, sys2, sys3, sys4, sys5, sys6, sys7, sys8, sys9])
ctl.bode_plot(response, initial_phase=0)
```
