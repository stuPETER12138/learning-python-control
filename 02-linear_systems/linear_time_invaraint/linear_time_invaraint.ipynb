Transfer Function

$$
G(s) = \frac{num(s)}{den(s)} = \frac{a_0 s^{m} + a_1 s^{m-1} + \dots + a_m}{b_0 s^{n} + b_1 s^{n-1} + \dots + b_n}
$$

```python
import control as ct
```

```python
num = [1, 0]
den = [2, 1, 2]

sys = ct.tf(num, den, name="sys", inputs=['u'], outputs=['v'])
```

```python
s = ct.tf('s')
```

```python
sys = 5 * (s + 10)/(s**2 + 2*s + 1)

val = sys(1 + 0.5j)

val
```

```python
sys = ct.tf([1, 2], [1, 2, 3], name='SISO transfer function')
response = ct.pole_zero_map(sys)
ct.pole_zero_plot(response)
```

```python
ct.root_locus_map(sys).plot()
```

```python
sys1 = ct.tf([1], [1, 2, 1], name='sys1')
sys2 = ct.tf([1, 0.2], [1, 1, 3, 1, 1], name='sys2')
ct.root_locus_plot([sys1, sys2], grid=False)
```
